```python
import pandas as pd
import seaborn as sns
```

The next cell loads the `diamonds` dataset from Seaborn

```python
data = sns.load_dataset('diamonds')
```

Display the first 5 rows of `data`

```python
data.head()
```

Sort `data` by ALL of the columns `cut`, `clarity`, and `color` columns. Then create a MultiIndex by setting the index of `data` to ALL `cut`, `clarity`, and `color`. Display the first 5 rows of `data`.

```python
data = data.sort_values(["cut","clarity", "color"])
data = data.set_index(["cut", "clarity", "color"])
data.head()
```

Group `data` by `cut` and `clarity` using the `mean` function, and save the result as a variable `grouped`. Display `grouped`.

```python
data.reset_index()
grouped = data.groupby(["cut","clarity"]).mean()
grouped
```

Display a cross-section of `grouped` using the `VVS1` clarity.

```python
grouped.xs("VVS1", level = "clarity").head()
```
