```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Load the data from CSV
data = pd.read_csv('HW2Data.csv')

# Display the first few rows of the dataframe
print(data)

# Linear Regression
X = data[['ture']]
y = data['Yield']
linModel = LinearRegression()
linReg = linModel.fit(X, y)

print(f"\nLinear Regression: \ny = {round(linReg.coef_[0],3)}x + {round(linReg.intercept_,3)}")

# Calculate r squared
r_sq_lin = linReg.score(X, y)
print(f"R^2: {r_sq_lin}")

# Make predictions
y_pred = linModel.predict(X)

# Evaluate the model
mse_lin = mean_squared_error(y, y_pred)
print(f'MSE: {mse_lin}')


# Polynomial Regression
poly_features = PolynomialFeatures(degree=2)
X_transformed = poly_features.fit_transform(X)
polyModel = LinearRegression()
polyReg = polyModel.fit(X_transformed, y)

print(f"\nPolynomial Regression: \ny = {round(polyReg.coef_[1],3)}x^2 + {round(polyReg.coef_[2],3)}x + {round(polyReg.intercept_,3)}")

# Calculate r squared
r_sq_poly = polyReg.score(X_transformed, y)
print(f"R^2: {r_sq_poly}")

# Make predictions
y_pred = polyModel.predict(X_transformed)

# Evaluate the model
mse_poly = mean_squared_error(y, y_pred)
print(f'MSE: {mse_poly}')
```

```python
import numpy as np

data = [12, 25, 7, 5, 10, 23, 5, 6, 27, 60, 13, 13, 10, 18, 5]

# Calculate mean
meanVal = np.mean(data)

# Calculate median
medianVal = np.median(data)

# Calculate count
countVal = len(data)

# Calculate min and max
minVal = np.min(data)
maxVal = np.max(data)

# Calculate range
rangeVal = maxVal - minVal

# Calculate standard deviation (sd) and variance
sdVal = np.std(data)
varianceVal = np.var(data)

# Calculate first quartile (Q1) and third quartile (Q3)
q1Val = np.percentile(data, 25)
q3Val = np.percentile(data, 75)

# Print the results
print(f'Mean: {meanVal}')
print(f'Median: {medianVal}')
print(f'Count: {countVal}')
print(f'Minimum: {minVal}')
print(f'Maximum: {maxVal}')
print(f'Range: {rangeVal}')
print(f'Standard Deviation (sd): {sdVal}')
print(f'Variance: {varianceVal}')
print(f'First Quartile (Q1): {q1Val}')
print(f'Third Quartile (Q3): {q3Val}')
```
