```python
import os

# Specify the directory you want to scan
directory = 'C:/Users/sfure/OneDrive - University of North Carolina at Chapel Hill/780 MACHINE LEARNING/780 Project/posters'

# Iterate over files in the directory
for filename in os.listdir(directory):
    # Check if '__' is in the filename
    if '__' in filename:
        # Create the new filename by replacing '__' with '_'
        new_filename = filename.replace('__', '_')
        
        # Define full paths
        old_filepath = os.path.join(directory, filename)
        new_filepath = os.path.join(directory, new_filename)
        
        
        if not os.path.exists(new_filepath):
            # Rename the file
            os.rename(old_filepath, new_filepath)
            print(f'Renamed: {filename} to {new_filename}')
```
